```python
import pandas as pd
import matplotlib.pyplot as plt

# Create a DataFrame from the log data

#df = pd.read_csv("D:\C_Code\serverless_genai\Fibonacci\GeneratorMethod\extracted_node.csv")
#df = pd.read_csv("D:\C_Code\serverless_genai\Fibonacci\GeneratorMethod\extracted_python.csv")
df = pd.read_csv("D:\C_Code\serverless_genai\Fibonacci\GeneratorMethod\extracted_ruby.csv")

# Convert Timestamp to datetime
df['Timestamp'] = pd.to_datetime(df['Timestamp'], unit='ms')

# Plotting the trends
fig, axs = plt.subplots(3, 1, figsize=(15, 25))

axs[0].plot(df['Timestamp'], df['Duration'], label='Duration', color='blue')
axs[0].set_title('Duration over Time')
axs[0].set_xlabel('Timestamp')
axs[0].set_ylabel('Duration (ms)')
axs[0].legend()

axs[1].plot(df['Timestamp'], df['Billed Duration'], label='Billed Duration', color='green')
axs[1].set_title('Billed Duration over Time')
axs[1].set_xlabel('Timestamp')
axs[1].set_ylabel('Billed Duration (ms)')
axs[1].legend()

axs[2].plot(df['Timestamp'], df['Max Memory Used'], label='Max Memory Used', color='red')
axs[2].set_title('Max Memory Used over Time')
axs[2].set_xlabel('Timestamp')
axs[2].set_ylabel('Max Memory Used (MB)')
axs[2].legend()

plt.tight_layout()
plt.show()
```
